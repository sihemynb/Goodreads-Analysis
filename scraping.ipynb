```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python
# Set up Selenium with ChromeDriver
options = Options()
options.headless = True  
driver = webdriver.Chrome(options=options)

# Create the URL to the mother link containing all the links to all years
url = "https://www.goodreads.com/list/show/143500"
driver.get(url)  # Request the URL to access the content

# not necessary, but it's good to wait a bit for the page to load 
# in case it takes too long to load etc
time.sleep(5)

years = [str(i)+"–"+str(i+9) for i in range(1980,2021,10)]
print(years)
urls=[]
year_links = {}

for year in years:
    # in the screen shots, we found that the links are in a elements (which means links)
    # and they have the text of the year. i'll use this to get the links
    # By.LINK_TEXT is a way to find link elements by their text
    # we search links with the text of the year written inside
    year_element = driver.find_element(By.LINK_TEXT, year) 
    urls.append(year_element.get_attribute('href'))
    year_links[year] = year_element.get_attribute('href')
    print(year_element.get_attribute('href'))

# close the browser after scraping
driver.quit()
```

```python
driver = webdriver.Chrome(options=options)
books_per_year = {} #dictionnary to contain the books by year

for year, url in year_links.items():
    driver.get(url)  # Request the URL to access the content
    time.sleep(4)

    books = []
    # like I did for the years, I inspected the page and found that 
    # the books have a class name of bookTitle
    book_elements = driver.find_elements(By.CLASS_NAME,'bookTitle')
    for book_element in book_elements[:100]: #only get the first 100 books
        books.append(book_element.text)
    books_per_year[year] = books
    

# close the browser after scraping
driver.quit()
```

```python
#recuperer les liens des livres
#2020-2029
import requests
from bs4 import BeautifulSoup

# URL de la page Goodreads
url = 'https://www.goodreads.com/list/show/143500'

# User-Agent pour éviter le blocage
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"
}

# Récupérer le contenu de la page
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# Scraper les liens des livres
books = soup.find_all("a", class_="bookTitle")

# Affichage des URLs
base_url = "https://www.goodreads.com"  # Base pour compléter les liens relatifs

for book in books:
    title = book.find("span").text.strip()
    href = book["href"]  # Récupération du href
    full_url = base_url + href  # Construction de l'URL complète
    book_links1 = [base_url + book["href"] for book in books]
    print(f"{title}: {full_url}")
print(book_links1)
```

```python
import requests
from bs4 import BeautifulSoup
import time
import re
import pickle

# User-Agent pour éviter le blocage
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"
}

# Fonction pour récupérer le nombre de pages
def get_num_pages(soup):
    pages_tag = soup.find("span", {"itemprop": "numberOfPages"})  # Vérifie d'abord l'ancienne méthode
    if pages_tag:
        return int(pages_tag.text.strip().split()[0])  # Extrait juste le nombre
    
    # Si non trouvé, vérifie les nouvelles balises possibles, fonction trouvée sur le github de MariaAntoniak
    pages_tag = soup.find("p", {"data-testid": "pagesFormat"}) or soup.find("div", {"data-testid": "contentContainer"})
    if pages_tag:
        match = re.search(r"\b\d+\b", pages_tag.text.strip())  # Extraction du nombre uniquement
        return int(match.group()) if match else "Non indiqué"

    return "Non indiqué"

def get_first_genre(soup):
    genre_tag = soup.find('a', {'class': 'actionLinkLite bookPageGenreLink'})  # Prend uniquement le premier
    return genre_tag.text.strip() if genre_tag else "Genre non trouvé"

# Liste pour stocker les données
book_data = []

# Scraper les informations pour chaque livre
for link in all_book_links :
    response = requests.get(link, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    # Trouver le titre
    title_tag = soup.find("h1", {"data-testid": "bookTitle"})
    title = title_tag.text.strip() if title_tag else "Titre non trouvé"

    # Trouver la date de publication (année uniquement)
    publication_tag = soup.find("p", {"data-testid": "publicationInfo"})
    if publication_tag:
        full_date = publication_tag.text.strip()
        match = re.search(r"\b\d{4}\b", full_date)
        publication_year = match.group() if match else "Année non trouvée"
    else:
        publication_year = "Année non trouvée"

    # Trouver le nombre de pages avec la fonction
    pages = get_num_pages(soup)

    # Trouver le nombre de ratings (lecteurs)
    ratings_tag = soup.find("span", {"data-testid": "ratingsCount"})
    ratings = re.sub(r"[^\d]", "", ratings_tag.text.strip()) if ratings_tag else "Non indiqué"
    
    # Trouver la note moyenne
    rating_tag = soup.find("div", {"class": "RatingStatistics__rating"})
    rating = rating_tag.text.strip() if rating_tag else "Non indiqué"

    # Trouver le premier genre
    # Trouver tous les éléments <a> avec la classe spécifique
    genre_links = soup.find_all('a', class_='Button Button--tag Button--medium')

    # Extraire le texte de chaque lien
    genres = [link.get_text() for link in genre_links]

    # Ajouter les données à la liste
    book_data.append({
        "Titre": title,
        "Année de publication": publication_year,
        "Pages": pages,
        "Ratings": ratings,
        "Note moyenne": rating,
        "Premier genre": genres
    })

    # Pause pour éviter le blocage
    time.sleep(1)

# Création du DataFrame
df = pd.DataFrame(book_data)

# Affichage du DataFrame
print(df)

# Sauvegarde en CSV si besoin
df.to_csv("goodreads_books.csv", index=False, encoding="utf-8")
```
